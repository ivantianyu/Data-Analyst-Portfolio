```python
# ==============================
# IMPORT LIBRARY
# ==============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Agar tampilan grafik lebih rapi
sns.set(style="whitegrid")

# ==============================
# MEMBACA DATASET
# ==============================

df = pd.read_csv("coffee_sales.csv")
print("üìä 5 Baris Pertama Dataset:")
print(df.head())
print("\n" + "="*60 + "\n")


# ==============================
# STRUKTUR DATA
# ==============================

print("üìã Informasi Struktur Data:")
df.info()
print("\n" + "="*60 + "\n")


# ==============================
# UKURAN DATASET
# ==============================

print(f"üìè Ukuran Dataset: {df.shape[0]} baris, {df.shape[1]} kolom")
print("\n" + "="*60 + "\n")


# ==============================
# STATISTIK DESKRIPTIF
# ==============================

print("üìà Statistik Deskriptif (Semua Kolom):")
print(df.describe(include="all"))
print("\n" + "="*60 + "\n")


# ==============================
# CEK MISSING VALUE
# ==============================

missing = df.isnull().sum()
print("üîç Missing Values per Kolom:")
print(missing)
if missing.sum() == 0:
    print("‚úÖ Tidak ada missing value.")
else:
    print("‚ö†Ô∏è Terdapat missing value.")
print("\n" + "="*60 + "\n")


# ==============================
# CEK DUPLIKASI DATA
# ==============================

duplicates = df.duplicated().sum()
print(f"üîÅ Jumlah Data Duplikat: {duplicates}")
if duplicates == 0:
    print("‚úÖ Tidak ada data duplikat.")
else:
    print("‚ö†Ô∏è Terdapat data duplikat.")
print("\n" + "="*60 + "\n")


# ==============================
# DETEKSI OUTLIER (IQR METHOD)
# ==============================

Q1 = df['money'].quantile(0.25)
Q3 = df['money'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df['money'] < Q1 - 1.5 * IQR) |
    (df['money'] > Q3 + 1.5 * IQR)
]

print(f"üö® Jumlah Outlier pada Kolom 'money': {len(outliers)}")
if len(outliers) == 0:
    print("‚úÖ Tidak ditemukan outlier berdasarkan metode IQR.")
else:
    print("Outlier yang ditemukan:")
    print(outliers)
print("\n" + "="*60 + "\n")


# ==============================
# FUNDAMENTAL VISUALIZATION
# ==============================

# --- BAR CHART: JUMLAH PENJUALAN PER JENIS KOPI ---
coffee_count = df['coffee_name'].value_counts()
print("‚òï Jumlah Penjualan per Jenis Kopi:")
print(coffee_count)
print("\nüí° Insight: Americano with Milk paling laris, diikuti Latte dan Americano.")
print("\n" + "="*60 + "\n")

plt.figure(figsize=(8,4))
coffee_count.plot(kind='bar', color='saddlebrown')
plt.title("Jumlah Penjualan per Jenis Kopi")
plt.xlabel("Jenis Kopi")
plt.ylabel("Jumlah Transaksi")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n" + "="*60 + "\n")


# --- LINE CHART: TOTAL PENJUALAN PER JAM ---
sales_by_hour = df.groupby('hour_of_day')['money'].sum()
print("‚è∞ Total Penjualan per Jam:")
print(sales_by_hour)
print("\nüí° Insight: Puncak penjualan terjadi jam 10 pagi dan 16 sore.")
print("\n" + "="*60 + "\n")

plt.figure(figsize=(8,4))
plt.plot(sales_by_hour.index, sales_by_hour.values, marker='o', color='darkgreen')
plt.title("Total Penjualan per Jam")
plt.xlabel("Jam")
plt.ylabel("Total Penjualan ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
print("\n" + "="*60 + "\n")


# --- SCATTER PLOT: HUBUNGAN JAM DAN HARGA ---
print("üìç Hubungan antara Jam dan Harga Minuman:")
print("Tidak ada tren linear yang jelas ‚Äî harga bervariasi tanpa pola kuat terhadap waktu.")
print("\nüí° Insight: Pelanggan membeli produk premium sepanjang hari.")
print("\n" + "="*60 + "\n")

plt.figure(figsize=(8,4))
plt.scatter(df['hour_of_day'], df['money'], color='navy', alpha=0.6)
plt.title("Hubungan Jam dan Harga")
plt.xlabel("Jam")
plt.ylabel("Harga ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
print("\n" + "="*60 + "\n")


# --- HISTOGRAM: DISTRIBUSI HARGA MINUMAN ---
unique_prices = sorted(df['money'].unique())
print("üí∞ Distribusi Harga Minuman:")
print(f"Harga unik: {unique_prices}")
print(f"Frekuensi: {[list(df['money']).count(p) for p in unique_prices]}")
print("\nüí° Insight: Mayoritas transaksi berada di tier premium ($35+).")
print("\n" + "="*60 + "\n")

plt.figure(figsize=(8,4))
plt.hist(df['money'], bins=15, color='teal', edgecolor='black')
plt.title("Distribusi Harga Minuman")
plt.xlabel("Harga ($)")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
print("\n" + "="*60 + "\n")


# ==============================
# ADVANCED VISUALIZATION
# ==============================

# --- BOXPLOT: DISTRIBUSI HARGA BERDASARKAN WAKTU ---
print("üåô Distribusi Harga Berdasarkan Waktu Hari:")
print(df.groupby('Time_of_Day')['money'].describe())
print("\nüí° Insight: Pelanggan paling boros di malam hari (rata-rata $32.89).")
print("\n" + "="*60 + "\n")

plt.figure(figsize=(8,4))
sns.boxplot(x='Time_of_Day', y='money', data=df, hue='Time_of_Day', legend=False)
plt.title("Distribusi Harga Berdasarkan Waktu")
plt.ylabel("Harga ($)")
plt.tight_layout()
plt.show()
print("\n" + "="*60 + "\n")


# --- PAIRPLOT MULTIVARIATE ---
print("üîÑ Pairplot: Hubungan antar variabel numerik utama")
print("Tidak ada cluster kuat ‚Äî distribusi cukup acak.")
print("\n" + "="*60 + "\n")

sns.pairplot(df[['hour_of_day','money','Weekdaysort']], diag_kind='kde')
plt.suptitle("Pairplot: hour_of_day, money, Weekdaysort", y=1.02)
plt.show()
print("\n" + "="*60 + "\n")


# ==============================
# EXPLORATORY DATA ANALYSIS (EDA)
# ==============================

# Univariate Analysis
print("üìä Analisis Univariate:")
print("Top 3 jenis kopi:", df['coffee_name'].value_counts().head(3).index.tolist())
print("Rentang harga: ${:.2f} ‚Äì ${:.2f}".format(df['money'].min(), df['money'].max()))
print("\n" + "="*60 + "\n")

# Bivariate Analysis
print("üìä Analisis Bivariate:")
cross_tab = pd.crosstab(df['coffee_name'], df['Time_of_Day'])
print("Cross-tabulation: coffee_name vs Time_of_Day")
print(cross_tab)
print("\n" + "="*60 + "\n")

# Multivariate Insight
print("üìä Analisis Multivariate:")
print("Rata-rata penjualan per waktu hari:")
print(df.groupby('Time_of_Day')['money'].mean().round(2))
print("\n" + "="*60 + "\n")


# ==============================
# ADVANCED EDA TECHNIQUES
# ==============================

# Grouping
print("üßÆ Grouping: Rata-rata Harga per Jenis Kopi")
avg_price = df.groupby('coffee_name')['money'].mean().sort_values(ascending=False)
print(avg_price.round(2))
print("\n" + "="*60 + "\n")

# Filtering
print("FilterWhere: Transaksi Afternoon dengan harga > $35")
df_filtered = df[(df['Time_of_Day'] == 'Afternoon') & (df['money'] > 35)]
print(f"Jumlah: {len(df_filtered)} transaksi")
print("\n" + "="*60 + "\n")

# Random Sampling
print("üé≤ Random Sampling (3 transaksi):")
sample = df.sample(3, random_state=42)
print(sample[['coffee_name', 'money', 'Time_of_Day', 'Date']])
print("\n" + "="*60 + "\n")

# Visualisasi Grouping
plt.figure(figsize=(8,4))
avg_price.plot(kind='bar', color='chocolate')
plt.title("Rata-rata Harga per Jenis Kopi")
plt.ylabel("Harga Rata-rata ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n" + "="*60 + "\n")


# ==============================
# TAHAPAN BARU: KORELASI SEMUA VARIABEL
# ==============================

print("üÜï TAHAPAN: KORELASI SEMUA VARIABEL (Termasuk Kategorikal)\n")

# 1. Buat kolom datetime
df['datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
print("‚úÖ Kolom datetime berhasil dibuat.")

# 2. Encode variabel kategorikal
df_encoded = df.copy()
categorical_cols = ['cash_type', 'coffee_name', 'Time_of_Day', 'Weekday', 'Month_name']
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

print("‚úÖ Variabel kategorikal berhasil di-encode ke numerik.")

# 3. Hitung korelasi semua variabel numerik (termasuk hasil encoding)
numeric_cols = df_encoded.select_dtypes(include=[np.number]).columns
corr_matrix = df_encoded[numeric_cols].corr()

# 4. Identifikasi korelasi kuat (|r| > 0.3)
threshold = 0.3
strong_corr = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        val = corr_matrix.iloc[i, j]
        if abs(val) > threshold:
            strong_corr.append((corr_matrix.columns[i], corr_matrix.columns[j], val))

print(f"\nüîç Pasangan variabel dengan korelasi kuat (|r| > {threshold}):")
if strong_corr:
    for var1, var2, r in strong_corr:
        print(f"  ‚Ä¢ {var1} ‚Üî {var2}: r = {r:.2f}")
else:
    print("  Tidak ada korelasi kuat ditemukan.")

# 5. Visualisasi heatmap lengkap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, cbar_kws={"shrink": .8})
plt.title("Heatmap Korelasi Semua Variabel (Termasuk Hasil Encoding)")
plt.tight_layout()
plt.show()

# 6. Rekomendasi untuk Advanced EDA
print("\nüéØ Rekomendasi untuk Advanced EDA Techniques:")
if strong_corr:
    print("Fokus pada pasangan berikut:")
    for var1, var2, r in strong_corr:
        print(f"  - Analisis '{var1}' vs '{var2}' menggunakan grouping atau cross-tabulation")
else:
    print("Tidak ada korelasi kuat. Lanjutkan dengan analisis berbasis business insight.")
print("\n" + "="*60 + "\n")
```
